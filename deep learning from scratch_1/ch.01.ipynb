 브로드캐스팅 기능으로 형상이 다른 행렬끼리도 원소 곱이 가능

```python
X = np.array([[51, 55],
             [14, 19],
             [0, 4]])

print(X)
```

```python
X.shape
```

```python
X[0]
```

```python
X[0][1]
```

```python
for row in X:
    print(row)
```

```python
X = X.flatten()
print(X)
```

flatten 은 1차원 배열로 전환(평탄화)

```python
X[np.array([0, 2, 4])]
```

```python
X > 15
```

```python
X[X > 15]
```

numpy 배열에 부등호 연산자를 사용한 결과는 boolean 배열.
이 boolean 배열을 사용해 배열 X에서 True에 해당하는 원소, 즉 값이 15보다 큰 원소만 꺼내는 게 가능

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(0, 6, 0.1)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

x = np.arange(0, 6, 0.1) ==> 0, 0.1, 0.2, ..., 5.8, 5.9 라는 데이터를 생성

```python
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label = 'sin')
plt.plot(x, y2, linestyle = '--', label = 'cos')
plt.xlabel('x')
plt.ylabel('y')
plt.title('sin & cos')
plt.legend()
plt.show()
```
