```python
import math
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['seaborn-whitegrid'])
```

```python
def linear_function(x):
    a = 0.5
    b = 2
    
    return a * x + b
```

```python
print(linear_function(5))
```

```python
linear_function(5)
```

```python
x = np.arange(-5, 5, 0.1)
y = linear_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Function')
plt.show()
```

```python
def qudratic_function(x):
    a = 1
    b = -1
    c = -2
    
    return a * x ** 2 + b * x + c
```

```python
print(qudratic_function(2))
```

```python
x = np.arange(-5, 5, 0.1)
y = qudratic_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Qudratic Function')
plt.show()
```

```python
def cubic_function(x):
    a = 4
    b = 0
    c = -1
    d = -8
    
    return a * x**3 + b * x**2 + c * x + d
```

```python
cubic_function(3)
```

```python
x = np.arange(-5, 5, 0.1)
y = cubic_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic Function')
plt.show()
```

```python
def my_func(x):
    a = 1
    b = -3
    c = 10
    
    return a * x**2 + b * x + c
```

```python
x = np.arange(-10, 10,0.1)
y = my_func(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(1.5, my_func(1.5))
plt.text(1.5-1.5, my_func(1.5) + 10, 'min value of f(x)\n({}, {})'.format(1.5, my_func(1.5)), fontsize = 10)
plt.title('my_func')
plt.show()
```

```python
min_val = min(y)
print(min_val)
```

```python
def get_minimum(x1, x2, f):
    x = np.arange(x1, x2, 0.01)
    y = f(x)
    
    plt.plot(x, y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('get_minimum')
    plt.show()
    
    return min(y)
```

```python
get_minimum(1, 4, my_func)
```

```python
def exponential_function(x):
    a = 4
    return a ** x
```

```python
exponential_function(4)
```

```python
exponential_function(0)
```

```python
x = np.arange(-3, 2, 0.1)
y = exponential_function(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1, 15)
plt.xlim(-4, 3)
plt.title('exponential_function')
plt.show()
```

```python
x = np.arange(-10, 10, 0.01)
y1 = -np.log(x)
y2 = -np.log(1-x)

plt.axvline(x = 0, color = 'k')
plt.axhline(y = 0, color = 'k')

plt.grid()
plt.plot(x, y1, 'b-', x, y2, 'r-')
plt.text(0.9, 2.0, 'y=-log(1-x)', fontsize = 15)
plt.text(0.1, 3, 'y=-log(x)', fontsize = 15)
plt.xlim(-0.3, 1.4)
plt.ylim(-0.5, 4)
plt.scatter(0.5, -np.log(0.5))
plt.show()
```
